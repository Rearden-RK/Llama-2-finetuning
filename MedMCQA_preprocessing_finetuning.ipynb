# Converting numerical values to the corresponding option

```python
# Mapping dictionary for option columns
option_columns = {1: 'opa', 2: 'opb', 3: 'opc', 4: 'opd'}

# Replace numerical values in 'cop' with corresponding option values
processed_medmcqa['cop'] = processed_medmcqa.apply(lambda row: row[option_columns[row['cop']]], axis=1)
```

```python
template = """
Here is a question with four options. Please choose the correct option.

### Question:
{question}

### Options:
A. {option_a}
B. {option_b}
C. {option_c}
D. {option_d}
"""
```

```python
#Creating the list
import pandas as pd

# Assuming medmcqa is your DataFrame
mcqs = []
for _, row in processed_medmcqa.iterrows():
  mcq = {
      "question": row['question'],
      "option_a": row['opa'],
      "option_b": row['opb'],
      "option_c": row['opc'],
      "option_d": row['opd'],
      "correct_option": row['cop'],
      "explanation": row['exp']
  }
  mcqs.append(mcq)
```

```python
mcqs[0]
```

```python
mcqs[0:10]
```

```python
dataset = []
for mcq in mcqs:
 prompt = template.format(**mcq)
 dataset.append({"input": prompt, "output": mcq["correct_option"], "explanation": mcq["explanation"]})
```

```python
# Specify the file path
file_path = r'C:\Users\reade\Downloads\data\dataset for finetuning\output.txt'

# Open the file in write mode
with open(file_path, 'w', encoding='utf-8') as file:
 # Iterate through each dictionary and write to the file
 for item in dataset:
     file.write(f"Input: {item['input']}\n")
     file.write(f"Output: {item['output']}\n")
     file.write(f"Explanation: {item['explanation']}\n")
     file.write("<s>\n\n") # Write the delimiter
```
